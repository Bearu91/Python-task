```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
sellers = pd.read_csv(r'C:\Users\pc\Desktop\python test\S_Data\sellers.csv')
products = pd.read_csv(r'C:\Users\pc\Desktop\python test\S_Data\products.csv')
order_items = pd.read_csv(r'C:\Users\pc\Desktop\python test\S_Data\order_items.csv')
orders = pd.read_csv(r'C:\Users\pc\Desktop\python test\S_Data\orders.csv')
product_category_name_translation = pd.read_csv(r'C:\Users\pc\Desktop\python test\S_Data\product_category_name_translation.csv')

# Merge datasets
merged_data = orders.merge(order_items, on='order_id', how='inner')
merged_data = merged_data.merge(products, on='product_id', how='inner')
merged_data = merged_data.merge(sellers, on='seller_id', how='inner')
merged_data = merged_data.merge(product_category_name_translation, on='product_category_name', how='inner')

# Convert date columns to datetime
merged_data['order_purchase_timestamp'] = pd.to_datetime(merged_data['order_purchase_timestamp'])

# Analysis of Sellers
seller_turnover = merged_data.groupby('seller_id')['price'].sum().reset_index()
seller_turnover.columns = ['seller_id', 'turnover']
top_sellers = seller_turnover.nlargest(10, 'turnover')
bottom_sellers = seller_turnover.nsmallest(10, 'turnover')

# Analysis of Products
product_turnover = merged_data.groupby('product_id')['price'].sum().reset_index()
product_turnover = product_turnover.merge(products, on='product_id', how='inner')
product_turnover = product_turnover.merge(product_category_name_translation, on='product_category_name', how='inner')
top_products = product_turnover.nlargest(10, 'price')
bottom_products = product_turnover.nsmallest(10, 'price')

# Turnover by Product Category
category_turnover = merged_data.groupby('product_category_name_english')['price'].sum().reset_index()
category_turnover.columns = ['product_category', 'turnover']
top_categories = category_turnover.nlargest(10, 'turnover')

# Product Weight vs Turnover and Price
weight_turnover_price = merged_data.groupby('product_id').agg({'price': 'sum', 'product_weight_g': 'mean'}).reset_index()
weight_turnover_price.columns = ['product_id', 'turnover', 'average_weight']
weight_turnover_price = weight_turnover_price.merge(products[['product_id', 'product_category_name']], on='product_id', how='inner')
weight_turnover_price = weight_turnover_price.merge(product_category_name_translation, on='product_category_name', how='inner')

# Segmentation of Sellers by Turnover
seller_turnover['segment'] = pd.qcut(seller_turnover['turnover'], 4, labels=['Low', 'Medium', 'High', 'Very High'])

# Segmentation of Products by Turnover
product_turnover['segment'] = pd.qcut(product_turnover['price'], 4, labels=['Low', 'Medium', 'High', 'Very High'])

# Visualizations

# Top 10 Sellers by Turnover
plt.figure(figsize=(10, 6))
sns.barplot(data=top_sellers, x='turnover', y='seller_id', hue='seller_id', palette='viridis', dodge=False, legend=False)
plt.title('Top 10 Sellers by Turnover')
plt.xlabel('Turnover')
plt.ylabel('Seller ID')
plt.show()

# Bottom 10 Sellers by Turnover
plt.figure(figsize=(10, 6))
sns.barplot(data=bottom_sellers, x='turnover', y='seller_id', hue='seller_id', palette='viridis', dodge=False, legend=False)
plt.title('Bottom 10 Sellers by Turnover')
plt.xlabel('Turnover')
plt.ylabel('Seller ID')
plt.show()

# Top 10 Products by Turnover
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='price', y='product_id', hue='product_id', palette='magma', dodge=False, legend=False)
plt.title('Top 10 Products by Turnover')
plt.xlabel('Turnover')
plt.ylabel('Product ID')
plt.show()

# Bottom 10 Products by Turnover
plt.figure(figsize=(10, 6))
sns.barplot(data=bottom_products, x='price', y='product_id', hue='product_id', palette='magma', dodge=False, legend=False)
plt.title('Bottom 10 Products by Turnover')
plt.xlabel('Turnover')
plt.ylabel('Product ID')
plt.show()

# Top 10 Product Categories by Turnover
plt.figure(figsize=(10, 6))
sns.barplot(data=top_categories, x='turnover', y='product_category', hue='product_category', palette='coolwarm', dodge=False, legend=False)
plt.title('Top 10 Product Categories by Turnover')
plt.xlabel('Turnover')
plt.ylabel('Product Category')
plt.show()

# Relationship between Product Weight, Turnover, and Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=weight_turnover_price, x='average_weight', y='turnover', hue='product_category_name_english', palette='tab10')
plt.title('Product Weight vs Turnover')
plt.xlabel('Average Product Weight (g)')
plt.ylabel('Turnover')
plt.show()

# Segmentation of Sellers by Turnover
plt.figure(figsize=(10, 6))
sns.countplot(data=seller_turnover, x='segment', hue='segment', palette='viridis', dodge=False, legend=False)
plt.title('Segmentation of Sellers by Turnover')
plt.xlabel('Segment')
plt.ylabel('Count of Sellers')
plt.show()

# Segmentation of Products by Turnover
plt.figure(figsize=(10, 6))
sns.countplot(data=product_turnover, x='segment', hue='segment', palette='magma', dodge=False, legend=False)
plt.title('Segmentation of Products by Turnover')
plt.xlabel('Segment')
plt.ylabel('Count of Products')
plt.show()

# Save the results to a notebook file
from IPython.display import display, Markdown
display(Markdown("### Top 10 Sellers by Turnover"))
display(top_sellers)
display(Markdown("### Bottom 10 Sellers by Turnover"))
display(bottom_sellers)
display(Markdown("### Top 10 Products by Turnover"))
display(top_products)
display(Markdown("### Bottom 10 Products by Turnover"))
display(bottom_products)
display(Markdown("### Top 10 Product Categories by Turnover"))
display(top_categories)
display(Markdown("### Product Weight vs Turnover and Price"))
display(weight_turnover_price)
display(Markdown("### Segmentation of Sellers by Turnover"))
display(seller_turnover)
display(Markdown("### Segmentation of Products by Turnover"))
display(product_turnover)
```
